```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
X = 6 * np.random.rand(200,1) - 3
y = 0.8 * X ** 2 + 0.9 * X + 2 + np.random.randn(200,1)
## y = 0.8X^2 + 0.9X + 2
```

```python
plt.scatter(X,y)
plt.plot()

# plt.plot(X,y,'b.')
# plt.show()
```

```python
# train test and split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2,test_size=0.2)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train,lr.predict(X_train),color='red')
plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures ## import for ploynomial regression 
```

```python
# ploy = PolynomialFeatures(degree=2,include_bias=False)
# [-1.71227616  2.93188965] it is nor return intercept(bias)
ploy = PolynomialFeatures(degree=2)

X_train_ploy = ploy.fit_transform(X_train)
X_test_ploy = ploy.transform(X_test)
```

```python
print(X_train[0])
print(X_train_ploy[0])
# [-1.71227616]
# [ 1.         -1.71227616  2.93188965]
## x^0 x^1 x^2
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train_ploy,y_train)
```

```python
y_pred = lr.predict(X_test_ploy)
```

```python
r2_score(y_test,y_pred)
```

```python
X_new = np.linspace(-3,3,200).reshape(200,1)
X_new_ploy = ploy.transform(X_new)
y_new = lr.predict(X_new_ploy)
```

```python
plt.plot(X_new,y_new,'r-',linewidth=2,label="Prediction")
plt.plot(X_train,y_train,'b.',label="Training point")
plt.plot(X_test,y_test,'g.',label="Testing point")
plt.legend()
plt.plot()
```
