```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
encoder = LabelEncoder()
df['species'] = encoder.fit_transform(df['species'])
```

```python
df.head()
```

```python
df = df[['sepal_length','petal_length','species']]
```

```python
df.head()
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# clf = LogisticRegression(multi_class='multinomial')
clf = LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
pd.DataFrame(confusion_matrix(y_test,y_pred))
```

```python
query = np.array([[3.4,2.7]]) ## this is prediction base on ("sepal_length","petal_length")
print("probability is : ",clf.predict_proba(query))
print("prediction is : ",clf.predict(query))
```

```python
## array([[7.25865651e-01, 2.73720718e-01, 4.13631316e-04]])
## this is probability of which flower is high prediction probability
```

```python
query = pd.DataFrame(
    [[3.4, 2.7]],
    columns=clf.feature_names_in_
)
print("probability is : ",clf.predict_proba(query))
print("prediction is : ",clf.predict(query))
```

```python
print(clf.n_features_in_)     # number of features
print(clf.feature_names_in_)   # names of features
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X.values,y.values,clf,legend=2)
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()

plt.title("Softmax Decision Boundary on Iris Dataset")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")

plt.legend(iris.target_names, title="Species")
plt.show()
```
